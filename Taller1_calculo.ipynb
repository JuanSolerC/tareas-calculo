</div>

\begin{align*}
\Huge f(x) = 2x + 1 \\
\Huge f(1), f(-2), f\left(\frac{1}{2}\right), f(a), f(-a), f(a + b)
\end{align*}




```python
from sympy import symbols, latex
from IPython.display import display, Math

# Definir la variable y la funci√≥n
x, a, b = symbols('x a b')
f = 2*x + 1

# Evaluar la funci√≥n en distintos valores
valores = [1, -2, 1/2, a, -a, a + b]
resultados = {v: f.subs(x, v) for v in valores}

# Mostrar la funci√≥n original
display(Math(r"f(x) = " + latex(f)))

# Mostrar cada evaluaci√≥n con formato matem√°tico correcto
for valor, resultado in resultados.items():
    expresion = f"f({latex(valor)}) = {latex(f.subs(x, valor))} = {latex(resultado)}"
    display(Math(expresion))
```

#### 2) Eval√∫e la funci√≥n definida por tramos en los valores indicados.

</div>


\begin{align*}
\Huge f(x) &= \begin{cases}
\Huge x^2 & \Huge \textrm{si}\, x < 0 \\
\Huge x+1 & \Huge \textrm{si}\, x \geq 0
\end{cases} \\
\Huge f(-2), f(-1), f(0), f(1), f(2)
\end{align*}




```python
# Definir la variable simb√≥lica
x = sp.symbols('x')

# Definir la funci√≥n por tramos
f = sp.Piecewise((x**2, x < 0), (x + 1, x >= 0))

# Evaluar la funci√≥n en los puntos dados
valores = [-2, -1, 0, 1, 2]
resultados = {x_val: f.subs(x, x_val) for x_val in valores}

# Mostrar los resultados en formato LaTeX
for x_val, resultado in resultados.items():
    display(Math(f"f({x_val}) = {resultado}"))

```

#### 4) Encuentre el dominio de la funci√≥n.

</div>

\begin{align*}
\Huge f(x) = \frac{x^4}{x^2 + x - 6} 
\end{align*}

```python
# Definir la variable simb√≥lica
x = symbols('x')

# Definir el denominador
denominador = x**2 + x - 6

# Encontrar los valores que hacen cero el denominador
restricciones = solve(Eq(denominador, 0), x)

# Crear la expresi√≥n en LaTeX para el dominio
dominio_latex = r"D_f = \mathbb{R} \setminus \{" + ", ".join(map(str, restricciones)) + r"\}"

# Mostrar el dominio en formato LaTeX
display(Math(dominio_latex))
```

*El dominio de ùëì(ùë•) son todos los n√∫meros reales excepto **x=‚àí3 y ùë•=2***

#### 5) Trace la gr√°fica de la funci√≥n haciendo primero una tabla de valores

</div>

\begin{align*}
\Huge f(x) = -x^2
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar la tabla de valores
valores_x = np.linspace(-3, 3, 7)  # 7 valores entre -3 y 3
valores_y = -valores_x**2

tabla = "| x  | f(x) = -x^2 |\n|----|-------------|\n"
for i in range(len(valores_x)):
    tabla += f"| {valores_x[i]:.1f} | {valores_y[i]:.1f} |\n"

display(Markdown(tabla))

# Graficar la funci√≥n
x_vals = np.linspace(-3, 3, 50)
y_vals = -x_vals**2

plt.figure(figsize=(6, 4))
plt.plot(x_vals, y_vals, label="$f(x) = -x^2$", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gr√°fica de f(x) = -x^2")
plt.show()
```

#### 9) Trace la gr√°fica de la funci√≥n definida por tramos.

</div>

\begin{align*}
\Huge f(x) &= \begin{cases}
\Huge 3 & \Huge \textrm{si}\, x < 2 \\
\Huge x-1 & \Huge \textrm{si}\, x \geq 2
\end{cases} \\
\end{align*}

```python
# Definir los valores de x en los diferentes tramos
x1 = np.linspace(-2, 1.99, 100)  # Tramo x < 2
y1 = np.full_like(x1, 3)         # f(x) = 3

x2 = np.linspace(2, 5, 100)      # Tramo x ‚â• 2
y2 = x2 - 1                      # f(x) = x - 1

# Graficar la funci√≥n por tramos
plt.figure(figsize=(6, 4))
plt.plot(x1, y1, label=r"$f(x) = 3$ si $x < 2$", color='blue')
plt.plot(x2, y2, label=r"$f(x) = x - 1$ si $x \geq 2$", color='red')

# Agregar punto vac√≠o y lleno en x = 2 para mostrar la discontinuidad
plt.scatter(2, 3, color='blue', edgecolors='black', s=50, label="Punto no incluido", zorder=3)
plt.scatter(2, 2-1, color='red', s=50, label="Punto incluido", zorder=3)

# Configuraci√≥n del gr√°fico
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Gr√°fica de la funci√≥n definida por tramos")

# Mostrar la gr√°fica
plt.show()
```

- *En una funci√≥n definida por tramos, **los puntos incluidos y no incluidos indican si un valor espec√≠fico de ùë•
pertenece o no a la funci√≥n en ese punto***
- *Punto vac√≠o en **(2,3)** ‚Üí porque la funci√≥n **no** toma ese valor en ùë•=2*
- *Punto lleno en **(2,1)** ‚Üí porque la funci√≥n **s√≠** toma ese valor en ùë•=2*

#### 10) En estos ejercicios se tiene que Trazar la gr√°fica de la funci√≥n, no localizando los puntos sino empezando con la gr√°fica de una funci√≥n est√°ndar y aplicando transformaciones

</div>

\begin{align*}
\Huge f(x) = x^2-1
\end{align*}

```python
# Definir las funciones
def g(x):
    return x**2  # Funci√≥n est√°ndar

def f(x):
    return x**2 - 1  # Funci√≥n transformada

# Generar valores de x
x_vals = np.linspace(-3, 3, 100)
g_vals = g(x_vals)
f_vals = f(x_vals)

# Crear la figura y los ejes
plt.figure(figsize=(6, 4))

# Graficar la funci√≥n est√°ndar g(x) = x^2
plt.plot(x_vals, g_vals, linestyle="dashed", color="gray", label=r"$g(x) = x^2$ (Funci√≥n base)")

# Graficar la funci√≥n transformada f(x) = x^2 - 1
plt.plot(x_vals, f_vals, color="blue", label=r"$f(x) = x^2 - 1$ (Transformada)")

# Marcar los puntos clave
plt.scatter([0], [-1], color='red', label="V√©rtice (0, -1)")
plt.scatter([-1, 1], [0, 0], color='green', label="Cortes con X (-1,0) y (1,0)")

# Agregar ejes
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle="--", linewidth=0.5)

# Etiquetas
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Transformaci√≥n de $g(x) = x^2$ a $f(x) = x^2 - 1$")
plt.legend()
plt.show()
```

#### 17) RESOLVER: encontrar las funciones f+g, f‚àíg, fg, f/g y determinar sus dominios.

</div>

\begin{align*}
\Huge f(x) = \sqrt{9 - x^2},  g(x) = \sqrt{x^2 -4}
\end{align*}

```python
# Definir la variable simb√≥lica
x = symbols('x')

# Definir las funciones f(x) y g(x)
f = sqrt(9 - x**2)
g = sqrt(x**2 - 4)

# Operaciones con funciones
f_plus_g = f + g
f_minus_g = f - g
f_times_g = f * g
f_div_g = f / g

# Mostrar las operaciones en LaTeX
display(Math(r"f + g = \sqrt{9 - x^2} + \sqrt{x^2 - 4}"))
display(Math(r"f - g = \sqrt{9 - x^2} - \sqrt{x^2 - 4}"))
display(Math(r"f \cdot g = \sqrt{(9 - x^2)(x^2 - 4)}"))
display(Math(r"\frac{f}{g} = " + f"{latex(f_div_g)}"))

# Determinar el dominio de f(x) y g(x)
dom_f = Interval(-3, 3)  # -3 ‚â§ x ‚â§ 3
dom_g = Interval(-S.Infinity, -2, True) | Interval(2, S.Infinity, True)  # (-‚àû, -2] U [2, ‚àû)

# Intersecci√≥n de los dominios para f+g, f-g, f¬∑g
dom_common = dom_f.intersect(dom_g)  # [-3,-2] U [2,3]

# Para f/g, quitamos los valores x = ¬±2 donde g(x) = 0
dom_f_div_g = dom_common - FiniteSet(-2, 2)  # [-3,-2) U (2,3]

# Mostrar los dominios en LaTeX
display(Math(r"\text{Dominio de } f + g \text{ y } f - g: " + f"{latex(dom_common)}"))
display(Math(r"\text{Dominio de } f \cdot g: " + f"{latex(dom_common)}"))
display(Math(r"\text{Dominio de } \frac{f}{g}: " + f"{latex(dom_f_div_g)}"))
```

#### 33) Trace un tri√°ngulo que tenga √°ngulo agudo Œ∏, y encuentre las otras cinco relaciones trigonom√©tricas de Œ∏

</div>

\begin{align*}
\Huge senŒ∏ = \frac{3}{5}
\end{align*}

```python
# Dado
sin_theta = Fraction(3, 5)

# Calculando las otras relaciones trigonom√©tricas
cos_theta = Fraction(math.sqrt(1 - sin_theta**2)).limit_denominator()
tan_theta = Fraction(sin_theta / cos_theta).limit_denominator()
cot_theta = Fraction(1 / tan_theta).limit_denominator()
sec_theta = Fraction(1 / cos_theta).limit_denominator()
csc_theta = Fraction(1 / sin_theta).limit_denominator()

# Calcular el lado adyacente usando el coseno
hipotenusa = 5
adyacente = hipotenusa * cos_theta

# Mostrar los resultados usando LaTeX
display(Math(r'\cos \theta = \frac{%s}{%s}' % (cos_theta.numerator, cos_theta.denominator)))
display(Math(r'\tan \theta = \frac{%s}{%s}' % (tan_theta.numerator, tan_theta.denominator)))
display(Math(r'\cot \theta = \frac{%s}{%s}' % (cot_theta.numerator, cot_theta.denominator)))
display(Math(r'\sec \theta = \frac{%s}{%s}' % (sec_theta.numerator, sec_theta.denominator)))
display(Math(r'\csc \theta = \frac{%s}{%s}' % (csc_theta.numerator, csc_theta.denominator)))

# Graficar el tri√°ngulo
plt.figure(figsize=(6, 6))
plt.plot([0, adyacente], [0, 0], color='blue', lw=2, label='Base')  # Base
plt.plot([0, 0], [0, 3], color='green', lw=2, label='Altura')  # Altura
plt.plot([0, adyacente], [3, 0], color='red', lw=2, label='Hipotenusa')   # Hipotenusa

# Etiquetas
plt.text(adyacente / 2, -0.5, f'{adyacente}', fontsize=12, ha='center')
plt.text(-0.5, 1.5, '3', fontsize=12, va='center')
plt.text(adyacente / 2, 2, '5', fontsize=12, ha='center')


# Agregar cuadr√≠cula
plt.grid(True)

plt.xlim(-1, 5)
plt.ylim(-1, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Tri√°ngulo con √°ngulo agudo Œ∏')
plt.legend()
plt.show()
```

#### 34) Trace un tri√°ngulo que tenga √°ngulo agudo Œ∏, y encuentre las otras cinco relaciones trigonom√©tricas de Œ∏

</div>

\begin{align*}
\Huge cosŒ∏ = \frac{9}{40}
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los lados del tri√°ngulo
hipotenusa = 40
adyacente = 9
opuesto = np.sqrt(hipotenusa**2 - adyacente**2)  # Calculado con Pit√°goras

# Coordenadas de los puntos del tri√°ngulo
A = (0, 0)
B = (adyacente, 0)
C = (adyacente, opuesto)

# Graficar el tri√°ngulo
fig, ax = plt.subplots()
ax.plot([A[0], B[0]], [A[1], B[1]], 'bo-', label=f'Adyacente ({adyacente})')
ax.plot([B[0], C[0]], [B[1], C[1]], 'go-', label=f'Opuesto ({opuesto:.2f})')
ax.plot([C[0], A[0]], [C[1], A[1]], 'ro-', label=f'Hipotenusa ({hipotenusa})')

# Etiquetas de los puntos
ax.text(A[0]-2, A[1], 'A (0,0)', fontsize=12, color='black')
ax.text(B[0]+1, B[1], f'B ({adyacente},0)', fontsize=12, color='black')
ax.text(C[0]+1, C[1], f'C ({adyacente},{opuesto:.2f})', fontsize=12, color='black')

# Mostrar √°ngulo Œ∏
angle = np.arccos(adyacente / hipotenusa)  # √Ångulo en radianes
arc = np.linspace(0, angle, 50)
x_arc = 2 * np.cos(arc)
y_arc = 2 * np.sin(arc)
ax.plot(x_arc, y_arc, 'purple')
ax.text(2, 1, r'$\theta$', fontsize=14, color='purple')

# Configuraci√≥n del gr√°fico
ax.set_xlim(-5, 45)
ax.set_ylim(-5, 45)
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_title("Tri√°ngulo Rect√°ngulo con Œ∏")
ax.legend()
ax.grid()

plt.show()

# Mostrar relaciones trigonom√©tricas en formato LaTeX
sin_theta = opuesto / hipotenusa
cos_theta = adyacente / hipotenusa
tan_theta = opuesto / adyacente
csc_theta = hipotenusa / opuesto
sec_theta = hipotenusa / adyacente
cot_theta = adyacente / opuesto

from IPython.display import display, Math

relations = f"""
\\begin{{aligned}}
\\sin(\\theta) &= {sin_theta:.2f} \\
\\cos(\\theta) &= {cos_theta:.2f} \\
\\tan(\\theta) &= {tan_theta:.2f} \\
\\csc(\\theta) &= {csc_theta:.2f} \\
\\sec(\\theta) &= {sec_theta:.2f} \\
\\cot(\\theta) &= {cot_theta:.2f} 
\\end{{aligned}}
"""
display(Math(relations))
```
